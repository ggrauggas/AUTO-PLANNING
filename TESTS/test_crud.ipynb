```python
import requests

BASE_URL = "http://localhost:4000/tasks"

def listar_tareas():
    response = requests.get(BASE_URL)
    tasks = response.json()
    if not tasks:
        print("üì≠ No hay tareas.")
        return
    for t in tasks:
        print(f"ID: {t['_id']}\nT√≠tulo: {t['title']}\nDescripci√≥n: {t['description']}\nPrioridad: {t['priority']}\nEstado: {t['status']}\n---")

def crear_tarea():
    title = input("T√≠tulo: ")
    description = input("Descripci√≥n: ")
    priority = input("Prioridad (low/medium/high): ").lower()
    status = input("Estado (pending/in-progress/completed): ").lower()
    task_data = {"title": title, "description": description, "priority": priority, "status": status}
    response = requests.post(BASE_URL, json=task_data)
    print("‚úÖ Tarea creada:", response.json())

def actualizar_tarea():
    task_id = input("ID de la tarea a actualizar: ")
    print("Deja vac√≠o si no quieres cambiar un campo.")
    title = input("Nuevo t√≠tulo: ")
    description = input("Nueva descripci√≥n: ")
    priority = input("Nueva prioridad (low/medium/high): ").lower()
    status = input("Nuevo estado (pending/in-progress/completed): ").lower()

    data = {}
    if title: data["title"] = title
    if description: data["description"] = description
    if priority: data["priority"] = priority
    if status: data["status"] = status

    response = requests.put(f"{BASE_URL}/{task_id}", json=data)
    print("‚úÖ Tarea actualizada:", response.json())

def eliminar_tarea():
    task_id = input("ID de la tarea a eliminar: ")
    response = requests.delete(f"{BASE_URL}/{task_id}")
    print("‚úÖ Resultado:", response.json())

def menu():
    while True:
        print("\n=== CRUD Tareas ===")
        print("1Ô∏è‚É£  Listar tareas")
        print("2Ô∏è‚É£  Crear tarea")
        print("3Ô∏è‚É£  Actualizar tarea")
        print("4Ô∏è‚É£  Eliminar tarea")
        print("5Ô∏è‚É£  Salir")
        opcion = input("Selecciona una opci√≥n: ")

        if opcion == "1":
            listar_tareas()
        elif opcion == "2":
            crear_tarea()
        elif opcion == "3":
            actualizar_tarea()
        elif opcion == "4":
            eliminar_tarea()
        elif opcion == "5":
            print("üëã Saliendo...")
            break
        else:
            print("‚ùå Opci√≥n inv√°lida, intenta de nuevo.")

if __name__ == "__main__":
    menu()
```
